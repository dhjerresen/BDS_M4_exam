# SGD Mechanics & Attention Context

## Part A: Manual SGD (First 3 Samples)

```python
import pandas as pd
# load dataset
data = pd.read_csv('https://raw.githubusercontent.com/aaubs/ds-master/main/data/Swedish_Auto_Insurance_dataset.csv')
```

```python
data.head()
```

```python
data = [
    (108, 392.5),
    (19, 46.2),
    (13, 15.7)
]

# Hyperparameters
alpha = 0.0001     # learning rate
w = 0.5            # initial weight

print("Sample | w(old) | x | y_hat | Loss | dL/dw | w(new)")
print("------------------------------------------------")

for i, (x, t) in enumerate(data, start=1):

    # Forward pass
    y_hat = x * w

    # Loss
    loss = (t - y_hat) ** 2

    # Gradient
    grad = 2 * x * (y_hat - t)

    # Weight update
    w_new = w - alpha * grad

    print(f"{i} | {w:.6f} | {x} | {y_hat:.6f} | {loss:.6f} | {grad:.6f} | {w_new:.6f}")

    # Update weight for next sample (SGD!)
    w = w_new
```

### Part A - Conclusion

The manual SGD calculations show how the weight is updated step by step based on the prediction error and gradient. Starting from an initial weight of 0.5, the parameter quickly adjusts toward better predictions, confirming that the learning rule moves the model in the direction that reduces the loss. The agreement between the manual results and the code output validates the correctness of the implementation and illustrates how gradient descent enables learning in a simple linear model.

## Part B: Attention Contextualization

### Homonym
**"fine"**

### Sentences
1. **Sentence 1:** *The police man gave me a fine*  
2. **Sentence 2:** *I am feeling fine*

```python
import numpy as np

# ========== Embeddings (2D vectors) ==========

embeddings = {
    "the": np.array([0.1, 0.1]),
    "police": np.array([0.9, 0.1]),
    "man": np.array([0.6, 0.2]),
    "gave": np.array([0.3, 0.3]),
    "me": np.array([0.1, 0.2]),
    "a": np.array([0.1, 0.1]),
    "fine": np.array([0.5, 0.5]),

    "i": np.array([0.1, 0.3]),
    "am": np.array([0.2, 0.3]),
    "feeling": np.array([0.1, 0.9]),
}
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=1, keepdims=True)

def self_attention(E):
    # Q = K = V = E
    scores = E @ E.T          # QK^T
    A = softmax(scores)      # attention weights
    context = A @ E          # new embeddings
    return context
```

```python
# Sentence 1 tokens
sentence1 = ["the", "police", "man", "gave", "me", "a", "fine"]

E1 = np.array([embeddings[word] for word in sentence1])

context1 = self_attention(E1)

# Find new embedding for "fine" in sentence 1
fine_index_1 = sentence1.index("fine")
fine_context_1 = context1[fine_index_1]

print("Fine embedding in sentence 1:", fine_context_1)


# Sentence 2 tokens
sentence2 = ["i", "am", "feeling", "fine"]

E2 = np.array([embeddings[word] for word in sentence2])

context2 = self_attention(E2)

# Find new embedding for "fine" in sentence 2
fine_index_2 = sentence2.index("fine")
fine_context_2 = context2[fine_index_2]

print("Fine embedding in sentence 2:", fine_context_2)
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

similarity = cosine_similarity(fine_context_1, fine_context_2)

print("\nCosine similarity between the two 'fine' embeddings:", similarity)
```

Lastly, I create a plot to vizualise the effect of attention

```python
import matplotlib.pyplot as plt
import numpy as np

# Combine embeddings
all_original = np.concatenate((E1, E2), axis=0)
all_context = np.concatenate((context1, context2), axis=0)

labels_original = sentence1 + sentence2
labels_context = sentence1 + sentence2

# Set up the plot
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# ======================
# Plot original embeddings
# ======================
for i, word in enumerate(labels_original):
    axes[0].scatter(all_original[i, 0], all_original[i, 1], label=word)

axes[0].set_title('Original Word Embeddings (Both Sentences)')
axes[0].set_xlabel('Dimension 1')
axes[0].set_ylabel('Dimension 2')
axes[0].grid(True)
axes[0].legend()

# ======================
# Plot attention outputs
# ======================
for i, word in enumerate(labels_context):
    axes[1].scatter(all_context[i, 0], all_context[i, 1], label=word)

axes[1].set_title('Self-Attention Output Vectors (Both Sentences)')
axes[1].set_xlabel('Dimension 1')
axes[1].set_ylabel('Dimension 2')
axes[1].grid(True)
axes[1].legend()

plt.show()
```

### Part B - Conclusion

This task demonstrates how self-attention enables contextualization of word meanings. Although the word *“fine”* starts with the same static embedding in both sentences, the attention mechanism adjusts its representation based on surrounding words. The resulting embeddings differ, and the cosine similarity below 1 confirms a semantic shift. This shows that self-attention captures context, allowing the model to distinguish between different meanings of the same word.
